```python
for i in range(5):
  for j in range(5):  # A
    if (j == 0 or j == 4) and i != 0 or (i == 0 or i == 2) and 0 < j < 4:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # B
    if j == 0 or (i == 0 or i == 2 or i == 4) and j < 4 or (j == 4 and i % 2 == 1):
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # C
    if i == 0 or i == 4 or j == 0:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # D
    if j == 0 or (i == 0 or i == 4) and j < 4 or (j == 4 and i != 0 and i != 4):
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # E
    if j == 0 or i == 0 or i == 2 or i == 4:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # F
    if j == 0 or i == 0 or i == 2:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # G
    if i == 0 or i == 4 or j == 0 or (i == 2 and j >= 2) or (j == 4 and i >= 2):
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # H
    if j == 0 or j == 4 or i == 2:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # I
    if i == 0 or i == 4 or j == 2:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # J
    if i == 0 or j == 2 or (i == 4 and j < 3) or (i == 3 and j == 0):
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # K
    if j == 0 or i + j == 2 or i - j == 2:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # L
    if j == 0 or i == 4:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # M
    if j == 0 or j == 4 or (i == j and j < 3) or (i + j == 4 and j > 1):
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print()

for i in range(5):
  for j in range(5):  # N
    if j == 0 or j == 4 or i == j:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # O
    if (i == 0 or i == 4) and 0 < j < 4 or (j == 0 or j == 4) and 0 < i < 4:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # P
    if j == 0 or (i == 0 or i == 2) and j < 4 or i == 1 and j == 4:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # Q
    if (i == 0 or i == 3) and 1 <= j < 4 or (j == 0 or j == 4) and 1 <= i < 3 or (i == 4 and j == 3) or (i == 2 and j == 2):
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # R
    if j == 0 or (i == 0 or i == 2) and j < 4 or i == 1 and j == 4 or i == 3 and j == 3 or i == 4 and j == 4:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # S
    if i == 0 and j > 0 or i == 2 and 0 < j < 4 or i == 4 and j < 4 or i == 1 and j == 0 or i == 3 and j == 4:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # T
    if i == 0 or j == 2:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print()

for i in range(5):
  for j in range(5):  # U
    if (j == 0 or j == 4) and i != 4 or i == 4 and 0 < j < 4:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # V
    if (j == 0 and i < 4) or (j == 4 and i < 4) or (i == 4 and j == 2):
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # W
    if j == 0 or j == 4 or (i == j and i >= 2) or (i + j == 4 and i >= 2):
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # X
    if i == j or i + j == 4:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # Y
    if (i == j and i < 3) or (i + j == 4 and i < 3) or (j == 2 and i >= 2):
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print(" ", end="")

  for j in range(5):  # Z
    if i == 0 or i == 4 or i + j == 4:
      print("*", end=" ")
    else:
      print(" ", end=" ")
  print()
```
